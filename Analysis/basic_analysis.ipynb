```python
import numpy as np
import pandas as pd
```

```python
delivery = pd.read_csv('..\Datasets\deliveries.csv')
```

```python
#accessing single column
delivery['batter']
```

```python
#accessing multiple columns
delivery[['batter','batsman_runs']]
```

```python
# Get only Virat Kohli's runs (single column)
kohli_runs = delivery.loc[
    delivery['batter'] == 'V Kohli',
    'batsman_runs'
]

kohli_runs
```

```python
# Get multiple details when Virat Kohli is batting
kohli_details = delivery.loc[
    delivery['batter'] == 'V Kohli',
    ['batter', 'non_striker', 'batsman_runs']
]

kohli_details
```

```python
# Get Virat Kohli's runs when AB de Villiers is the non-striker
kohli_with_ab = delivery.loc[
    (delivery['batter'] == 'V Kohli') & 
    (delivery['non_striker'] == 'AB de Villiers'),
    ['batter', 'non_striker', 'batsman_runs']
]

kohli_with_ab
```

```python
# Get all batters who have ever played for RCB
rcb_batters = delivery[delivery['batting_team'] == 'Royal Challengers Bangalore']

rcb_batters['batter'].unique()
```
