```python
import numpy as np
import pandas as pd
```

```python
delivery = pd.read_csv('..\Datasets\deliveries.csv')
```

```python
batting_performance = delivery.groupby(['batter'])['batsman_runs'].sum().reset_index()
batting_performance.rename(columns={'batsman_runs':'total_runs'},inplace=True)
```

```python
#Top 5 batters based on total runs
batting_performance.sort_values(by='total_runs',ascending=False).head(5)
```

```python
bowling_performance = delivery.groupby(['bowler'])['total_runs'].sum().reset_index()
bowling_performance.rename(columns={'total_runs':'runs_conceded'},inplace=True)
```

```python
#Top 5 bowlers

bowling_performance.sort_values(by='runs_conceded',ascending=True).head(5)
```
